## the Breast Cancer dataset and the reasons behinde it as another dataset 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import seaborn as sns
sns.set_style('darkgrid')

%matplotlib inline
```

```python
#importing dataset
df1 = pd.read_csv("data.csv")
```

```python
#importing dataset
df2 = pd.read_csv("Breast Cancer METABRIC.csv")
```

```python
df = pd.concat([df1,df2], axis=0)
##df = df.rename(columns={"Unnamed: 1": "weekday"})
##df = df.dropna(subset=["weekday"])
df.head()
```

```python
plt.figure(figsize = (20, 10))
matrix = np.triu(df.corr())
sb.heatmap(df.corr(), annot = True, linewidth = 0.9, mask = matrix, cmap = "BrBG")
```

```python
#Second method: plotting using barplot
x = df['diagnosis'].value_counts().index
y = df['diagnosis'].value_counts().values

plt.figure(figsize=(9,6))
ax= sns.barplot(x, y,data= df ,alpha=0.8 )
plt.title("Cancer results")
plt.xlabel('', fontsize=12);
```

```python
print(f'METABRIC Dataset Shape: {df.shape}')
print(f'METABRIC Dataset Memory Usage: {df.memory_usage().sum() / 1024 ** 2:.2f} MB')
```
